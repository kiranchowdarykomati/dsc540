```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
df_all = pd.read_csv('population_by_country_2020.csv')
df=df_all.head(20)
```

```python
# Scatter plot
plt.scatter(df['Med. Age'],df['Fert. Rate'])
plt.title('Median Age Vs Fertility')
plt.xlabel('Median Age')
plt.ylabel('Fertility')
plt.show()
```

```python
# Histogram
plt.hist(df['Med. Age'],bins=20)
plt.show()
```

```python
# Bar Chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df['Country (or dependency)'],df['Population (2020)'])
plt.xticks(rotation=90)
plt.show()
```
