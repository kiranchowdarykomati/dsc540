```python
import pandas as pd
import numpy as np
%matplotlib inline
from numpy import nan as NA
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Load data
metArtdf = pd.read_csv('MetObjects.csv')
pd.options.display.max_columns = 100
metArtdf
```

```python
# Dropping any rows and columns that are all NAs
metArtdf1=metArtdf.dropna(how='all',axis=1)
metArtdf1
```

```python
# Delete duplicate rows
metArtdf1.drop_duplicates()
```

```python
# Get Subset with first three columns 
metArtdf_subset1=metArtdf[['Object Number','Is Highlight','Is Public Domain']]
metArtdf_subset1
```

```python
# Get Subset with 1,4,5 columns
metArtdf_subset2=metArtdf[['Object Number','Object ID','Department']]
metArtdf_subset2
```

```python
# Merge subsets on the key "Object Number"
merged_df=pd.merge(metArtdf_subset1, metArtdf_subset2, on='Object Number')
merged_df
```

```python
# Pivot columns to rows
merged_df.stack()
```

```python
#Applying Cross tab to calculate group frequencies
pd.crosstab(metArtdf1['Artist Role'],metArtdf1['Artist Nationality'], margins=True)
```
